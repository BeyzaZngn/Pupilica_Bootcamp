```python
def daireAlaniHesapla(r):
  """
  Description: Bu fonksiyon dairenin alanini hesaplar.
  input parameters: r
  """
  pi = 3.14
  alan = pi * (r ** 2)
  print(f"Daire Alani: {alan}")
```

```python
daireAlaniHesapla(5)
```

```python
def daireAlaniHesapla(r = 6):
  """
  Ben bir yarıçap vermesem bile default olarak 6 alacak.
  Description: Bu fonksiyon dairenin alanini hesaplar.
  input parameters: r
  output: dairenin alani (alan)
  """
  PI = 3.14  # Değeri ASLA değişmeyenler böyle tanımlanabilir.
  alan = PI * (r ** 2)
  print(f"Daire Alani: {alan}")
  return alan # Dışarıya çıkarmak istediğim ifadeyi return ile yazıyoruz.
```

```python
daireAlaniHesapla()
```

```python
user_input = input("Bir daire yaricapi giriniz: ")
print(f"girilen deger: {user_input} ve bu degerin tipi: {type(user_input)}")
daireAlaniHesapla(int(user_input))
```

```python
# dairenin cevresi hesaplama: 2*pi*r
def dairenin_cevresi(r):
  pi = 3.14
  cevre = 2 * pi * r
  return cevre
```

```python
cevre_input = input("Bir daire yaricapi giriniz: ")
cevre_hesaplanan = dairenin_cevresi(int(cevre_input))
print(cevre_hesaplanan)
```
